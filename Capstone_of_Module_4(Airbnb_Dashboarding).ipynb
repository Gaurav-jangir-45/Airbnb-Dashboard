## Data Wrangling ##


```python
# Making the copy of the original dataframe
df = Airbnb_df.copy()
# Making a copy of the dataset is important to ensure that our original dataset remains intact and safe from any modifications.
```

```python
# Droping 'neighbourhood_group' column because it has 100% missing values
df.drop('neighbourhood_group', axis=1, inplace=True)
```

```python
# Check if number_of_reviews == 0
len(df[df['number_of_reviews'] == 0])
```

```python
# Dealing with Null values in reviews_per_month, last_review, licnese, price.

'''Since number_of_reviews is ralated to reviews_per_month and last_review
(when number_of_reviews are 0 then related rows of reviews_per_month and last_review are null)
so we will use 0.0 for reviews_per_month and pd.NaT for last_review.'''

# To fill missing values in 'reviews_per_month' with '0.0'
df.loc[df['number_of_reviews'] == 0, ['reviews_per_month']] = [0.0]

# To fill missing values in 'last_review' with 'pd.NaT'
df.loc[df['number_of_reviews'] == 0, ['last_review']] = [pd.NaT]

# To fill missing values in 'licnese' column with 'Not Available'
df['license'].fillna('Not Available', inplace=True)

# To fill missing values in 'price' column with 'Mean' price
mean_price = df['price'].mean() # Calculate mean price
df['price'].fillna(mean_price, inplace=True) # Fill missing values with mean price
```

```python
# For checking my changes using .info() function
df.info()
```

```python
# Extract month from 'last_review' column
df['month'] = df['last_review'].dt.month

# Create a function to determine the season
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
# Extract month from 'last_review' column
df['month'] = df['last_review'].dt.month

# Create a function to determine the season
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'


# Apply the function to create a 'season' column
df['season'] = df['month'].apply(get_season)

# Group by season to aggregate the data
seasonal_bookings = df.groupby('season').agg({
    'number_of_reviews': 'sum',  # Sum of reviews for seasonal bookings
    'price': 'mean'  # Average price for the season
}).reset_index()

# Display the results
print("\nSeasonal Bookings:")
print(seasonal_bookings)
# Apply the function to create a 'season' column
df['season'] = df['month'].apply(get_season)

# Group by season to aggregate the data
seasonal_bookings = df.groupby('season').agg({
    'number_of_reviews': 'sum',  # Sum of reviews for seasonal bookings
    'price': 'mean'  # Average price for the season
}).reset_index()

# Step 5: Display the results
print("\nSeasonal Bookings:")
print(seasonal_bookings)

```

```python
# Droping 'month' column
df.drop('month', axis=1, inplace=True)
```

```python
# To 'price_categoty' column in dataset

# Define a function to get price_category
def get_price_category(price):
    if price <= 100:
        return 'Budget'
    elif 100 < price <= 200:
        return 'Standard'
    elif 200 < price <= 500:
        return 'Premium'
    else:
        return 'Luxury'

# Apply the function to create a 'property_size_category' column
df['price_category'] = df['price'].apply(get_price_category)

# Display the updated DataFrame with the new column
print(df[['price', 'price_category']].head())
```

```python
# To add 'Property_size' column in dataset

# Define a function to categorize property size based on room type
def get_property_size(room_type):
    if room_type == 'Entire home/apt':
        return 'Large'  # Larger space, typically an entire home or apartment
    elif room_type == 'Private room':
        return 'Medium'  # Private room is typically medium-sized
    elif room_type == 'Shared room':
        return 'Small'  # Shared rooms are usually smaller
    elif room_type == 'Hotel room':
        return 'Small'  # Hotel rooms are usually smaller
    else:
        return 'Unknown'  # If room type is unknown

# Apply the function to create a 'property_size' column
df['property_size'] = df['room_type'].apply(get_property_size)

# Display the updated DataFrame with the new column
print(df[['room_type', 'property_size']].head())
```

```python
# Define a function to categorize host experience level based on the number of listings
def get_host_experience_level(listings_count):
    if listings_count <= 10:
        return 'Beginner'  # Hosts with few listings are considered beginners
    elif 10 <= listings_count <= 30:
        return 'Intermediate'  # Hosts with a moderate number of listings
    else:
        return 'Expert'  # Hosts with many listings are considered experts

# Apply the function to create a 'host_experience_level' column
df['host_experience_level'] = df['calculated_host_listings_count'].apply(get_host_experience_level)

# Display the updated DataFrame with the new column
print(df[['calculated_host_listings_count', 'host_experience_level']].head())
```

```python
# To add 'Host_reponse_rate' column in dataset

# Calculate host response rate
df['host_response_rate'] = df.apply(
    lambda row: row['reviews_per_month'] / row['number_of_reviews'] if row['number_of_reviews'] != 0 else 0,
    axis=1
)# Avoid division by zero by using a conditional expression

# Display the updated DataFrame with the new column
print(df[['reviews_per_month', 'number_of_reviews', 'host_response_rate']].head())
```
